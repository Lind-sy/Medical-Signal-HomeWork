```python
import matplotlib.pyplot as plt
import sounddevice as sd
import numpy as np
%matplotlib inline

fs = 44100  # Hertz
duration = 5  # seconds

mydata = sd.rec(int(fs * duration), samplerate=fs, channels=2)
sd.wait()
plt.plot(mydata)
myRecording = sd.playrec(mydata,fs,channels=2)
```

```python
whiteNoise = np.random.normal(0,0.1,mydata.shape)
signal = mydata + whiteNoise
plt.plot(signal)
sd.playrec(signal,fs,channels=2)
```

```python
inverseS = -mydata
plt.plot(inverseS)
sd.playrec(inverseS,fs,channels=2)
```
