```python
import numpy as np
from numpy.matlib import randn
from scipy.signal import sawtooth
import matplotlib.pyplot as plt
import pandas as pd
from scipy.ndimage import uniform_filter
```

```python
A =1; f=5; Fs =1000; phi = 0; ti = 0; tf = 0.5
t = np.arange(ti,tf-1/Fs,1/Fs)
s =  A * sawtooth(2 * np.pi * f * t + phi)
s_n = s+1*randn(1,len(s))
order = 4
df = pd.DataFrame(s_n.T)
M = df.rolling(window=order).mean()
plt.plot(M)
plt.margins(0.1, 0.1)
plt.grid(True)
plt.show()
```

```python
def moving_average(s, order):
    window= np.ones(int(order))/float(order)
    x = np.convolve(s, window, 'same')
    return x
```

```python
N=10
order = 3
M2 = moving_average(np.ravel(s_n),N)
plt.plot(M2)
plt.margins(0.1, 0.1)
plt.grid(True)
plt.show()
```
