```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
Fs = 100 ; ti = 0 ; tf = 0.5 
t  = np.arange(ti,tf,1/Fs)
f = 5 
A = 10  
phi = 0 

# simulation of a sinusoidal signal s(t) at continuous time : s_c
s_c = A * np.sin( 2*np.pi*f*t + phi ) 
phi = np.pi/2
s_c2 = A * np.sin( 2*np.pi*f*t + phi ) 
phi = np.pi
s_c3 = A * np.sin( 2*np.pi*f*t + phi ) 
plt.plot(t,s_c)
plt.plot(t,s_c2)
plt.plot(t,s_c3)
plt.xlabel('continuous time')
plt.ylabel('Amplitude')
plt.title('continuous sinusoidal signal with different phase')
```

```python
n  = np.arange(0,(len(t)-1),1) 
# sinusoidal signal sinusoidal s(t) at discrete time : s_d
s_d = A * np.sin( 2*np.pi*f/Fs*n + phi ) 

samples_per_cycle = Fs/f

plt.plot(s_d)
plt.stem(s_d)
plt.xlabel('discrete time')
plt.ylabel('Amplitude')
plt.title('discrete sinusoidal signal')
```

```python
decim = 5#decimation factor
F_nq = Fs/2 #minimal Sampling rate - Nyquist Frquency
t_dec = signal.decimate(t,decim)
s_dec = signal.decimate(s_c,decim)
#print(len(s_dec))
Fs_dec = len(s_dec) * 2
samples_per_cycle = Fs_dec/f

plt.plot(t_dec,s_dec)
plt.xlabel('continuous time')
plt.ylabel('Amplitude')
plt.title('decimated continuous sinusoidal signal')
```
