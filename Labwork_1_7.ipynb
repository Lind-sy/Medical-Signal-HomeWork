```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
def indices(list, filtr=lambda x: bool(x)):
    return [i for i,x in enumerate(list) if filtr(x)]
#Create triangular pulse
def trigen(t, Tw, skew):
    Ta=float(Tw)/float(2*skew)
    y = [0 for x in range(len(t))] 
    index1 = indices(t, lambda x: x>(-Tw/2) and x<Ta)
    index2 = indices(t, lambda x: x<(-Tw/2) and x>Ta)
    index3 = indices(t, lambda x: x==Ta) 
    for i in index1:
        rez = (2.0*t[i]+Tw)/(Tw*(1.0+skew))
        y.insert(i, rez)
    for i in index2:
        rez = (2.0*t[i]+Tw)/(Tw*(1.0+skew))
        y.insert(i, rez)
    for i in index3:
        y[i] = 1.0
    return y
```

```python
fs = 1
t = np.arange(-10,9,1/fs)
w = 10
x = trigen(t, w, 1) #trinagular pulse function 
plt.plot(x)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
```

```python
d = signal.unit_impulse(28,[9,12,20]) #dirac delta function
plt.stem(d)
plt.margins(0.1, 0.1)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
y1 = np.convolve(x,d) #performe time shifting between the signals
plt.plot(y1)
plt.xlabel('Time [samples]')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
y2 = x*d #performe sampling effect
plt.stem(y2)
```
